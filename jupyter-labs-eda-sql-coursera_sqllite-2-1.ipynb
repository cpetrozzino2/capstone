### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
#Open connection and cursor
conn = sqlite3.connect('my_data1.db')
cursor = conn.cursor()

#Display distince launch sites
cursor.execute("SELECT DISTINCT Launch_site FROM SPACEXTBL")
results=cursor.fetchall()
for row in results:
    print(row)
    
    
#Close connection and cursor
cursor.close()
conn.close()
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
#Open connection and cursor
conn = sqlite3.connect('my_data1.db')
cursor = conn.cursor()

#Display
fetch_num = 5
cursor.execute("SELECT * FROM SPACEXTBL WHERE Launch_site Like 'CCA%' LIMIT ?",(fetch_num,))
results=cursor.fetchall()
for row in results:
    print(row)

#Close cursor and connection
cursor.close()
conn.close()
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
#Open connection and curssor
conn = sqlite3.connect('my_data1.db')
cursor = conn.cursor()

#Get sum of payload_mass
cursor.execute("SELECT SUM(PAYLOAD_MASS__KG_) FROM SPACEXTBL WHERE Customer = 'NASA (CRS)'")
sum_value=cursor.fetchone()[0]
print ("Payload_mass total is ", sum_value)
       
#Close cursor and connection
cursor.close()
conn.close()
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
#Open connection and curssor
conn = sqlite3.connect('my_data1.db')
cursor = conn.cursor()

#Get average of payload_mass
cursor.execute("SELECT AVG(PAYLOAD_MASS__KG_) FROM SPACEXTBL WHERE Booster_Version Like 'F9 v1.1%'")

average_pm = cursor.fetchone()[0]
print ("average_pm is ", average_pm)
       
#Close cursor and connection
cursor.close()
conn.close()
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
#Open connection and curssor
conn = sqlite3.connect('my_data1.db')
cursor = conn.cursor()

#Get the names of the boosters
cursor.execute("SELECT DISTINCT(Booster_version) FROM SPACEXTBL WHERE (Landing_Outcome = 'Success (drone ship)') AND (PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000)")
results=cursor.fetchall()

for row in results:
    print(row)
    
#Close cursor and connection
cursor.close()
conn.close()
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
#Open connection and curssor
conn = sqlite3.connect('my_data1.db')
cursor = conn.cursor()

#Get the names of the boosters
cursor.execute("SELECT DISTINCT(Booster_Version) FROM SPACEXTBL WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) from SPACEXTBL)")
results=cursor.fetchall()

for row in results:
    print(row)
    
#Close cursor and connection
cursor.close()
conn.close()
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
import calendar
import datetime
```

```python
#Open connection and curssor
conn = sqlite3.connect('my_data1.db')
cursor = conn.cursor()

#Get the names of the boosters
cursor.execute("SELECT substr(Date, 4,2), Landing_Outcome, Booster_Version, Launch_Site FROM SPACEXTBL WHERE (substr(Date, 7,4) = '2015') and (Landing_Outcome = 'Failure (drone ship)')")
results=cursor.fetchall()

print("2015 Drone Ship Landing Failures\n")
for row in results:
    print(row)
    
print('\n')

for row in results:
    month_str = (row[0])
    month_int = int(month_str)
    month_name = calendar.month_name[month_int]
    print(month_name,row[1], row[2],row[3])

    
#Close cursor and connection
cursor.close()
conn.close()
```

### Task 10




##### Rank the  count of  successful landing_outcomes between the date 04-06-2010 and 20-03-2017 in descending order.


```python
#Open connection and curssor
conn = sqlite3.connect('my_data1.db')
cursor = conn.cursor()

#Get the names of the boosters

cursor.execute("SELECT substr(Date,7,4), COUNT(*) AS year_count from SPACEXTBL WHERE Landing_Outcome Like 'Success%' GROUP BY substr(Date,7,4) ORDER BY year_count DESC")
               
results=cursor.fetchall()

for row in results:
    print(row)
    
#Close cursor and connection
cursor.close()
conn.close()
```
