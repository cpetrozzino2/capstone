```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
accuracy_data={'model_type':['logreg','svm','tree','knn'],
                'value':[0.85, 0.85, .875, .85]}
               
df=pd.DataFrame(accuracy_data)
sns.barplot(data=df,x='model_type',y='value')
plt.ylabel("accuracy", fontsize=20)
plt.xlabel("model type",fontsize=20)
plt.title("accuracy based on training data")
plt.show()
```

```python
score_data={'model_type':['logreg','svm','tree','knn'],
                'value':[0.83, 0.83, .78, .83]}
               
df_score=pd.DataFrame(score_data)
sns.barplot(data=df_score,x='model_type',y='value')
plt.ylabel("accuracy", fontsize=20)
plt.xlabel("model type",fontsize=20)
plt.title("accuracy based on testing data")
plt.show()
```
